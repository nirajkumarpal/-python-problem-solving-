```python
# use for loop to iterate from 0 to 100 and print the sum of all even numbers and the sum of all odd numbers
even = 0
odd = 0
for i in range(101):
    if i % 2 == 0:
        even += i

    else:
        odd += i   


print('The sum of all even is ',even)
print('The sum of all odd is ',odd)
        
```
